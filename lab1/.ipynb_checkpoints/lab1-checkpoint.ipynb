```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
 
#Convert data to Panda package
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])

# 1. Please view the data and refer to the specific information.

print(iris)
```

```python
# 2. Display the number of rows and columns

num_rows = len(iris.index)
print('Number of rows is {}'.format(num_rows))
num_col = iris.shape[1]
print('Number of columns is {}'.format(num_col))
```

```python
# 3. View basic information for individual columns. Use the 'describe' and 'groupby' methods

iris.describe()
```

```python
iris.groupby('petal width (cm)').describe()
```

```python
# 4. Using the 'head' method please display the first 5 rows

print(iris.head())
```

```python
# 5. Using “dropna” check if the database contains missing data

print(iris.count())
print(iris.dropna().count())
```

```python
# 6. Sort the data in ascending order relative to the second column.

iris_sorted = iris.sort_values('sepal width (cm)', axis=0, ascending=True)
print(iris_sorted)
```

```python
# 7. Find the minimum and maximum petal length (column 3 - petal length). Give the index of each of these values.

min_petal_length = iris['petal length (cm)'].min()
max_petal_length = iris['petal length (cm)'].max()
print(iris[min_petal_length == iris['petal length (cm)']][['petal length (cm)']])
print(iris[max_petal_length == iris['petal length (cm)']][['petal length (cm)']])
```

```python
# 8. Calculate the standar deviation for each column.

print(iris.std())
```

```python
# 9. Identify rows for which the length of the flower cup (sepal length) is greater than the average length (the whole set)

iris_sepal_length_gt_avg = iris[iris['sepal length (cm)'] > iris['sepal length (cm)'].mean()].index
print(iris_sepal_length_gt_avg)

# To get info based on this Index object you can simply use below print
print(iris.iloc[iris_sepal_length_gt_avg])
```

```python
# 10. Display histograms for individual parameters with regard to belonging to a given class.

print(iris.hist(figsize=(15,15)))  # default backend call is plt.hist
```
