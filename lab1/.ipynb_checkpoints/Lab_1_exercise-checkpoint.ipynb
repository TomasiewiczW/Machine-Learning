# 7. Find the minimum and maximum petal length (column 3 - petal length). Give the index of each of these values.

```python
iris['petal length (cm)'].max() # ordinary way of finding max
```

```python
iris.sort_values(by = 'petal length (cm)', ascending=True)['petal length (cm)']
# This operation creates a series object.
```

```python
iris_petal_len = iris.sort_values(by = 'petal length (cm)', ascending=True)['petal length (cm)']
```

```python
# Finding maximum indexies:

arr_max_indexies = []
for i in range(iris_petal_len.size):
    if (max(iris_petal_len.values) == iris_petal_len.values[i]):
        arr_max_indexies.append(iris_petal_len.index[i])

arr_max_indexies

# Not very elegant, but it works. In any case.
```

```python
# Finding minimum indexies:

arr_min_indexies = []
for i in range(iris_petal_len.size):
    if (min(iris_petal_len.values) == iris_petal_len.values[i]):
        arr_min_indexies.append(iris_petal_len.index[i])

arr_min_indexies

# Not very elegant, but it works. In any case.
```

# 9. Identify rows for which the length of the flower cup (sepal length) is greater than the average length (the whole set)

```python
iris_sepal_len = iris.sort_values(by = 'sepal length (cm)', ascending=True)['sepal length (cm)']
```

```python
arr_greater_than_mean_indexies = []
for i in range(iris_sepal_len.size):
    if (iris_sepal_len.mean() < iris_sepal_len.values[i]):
        arr_greater_than_mean_indexies.append(iris_sepal_len.index[i])

arr_greater_than_mean_indexies
```
